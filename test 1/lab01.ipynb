### Q2: What is the value of the predictive linear function $f_w(x)$ for $x=-0.15$ with $w_0=-0.1, w_1=0.2$?

☐ $f$ = 


```python
#YO# predictive function definition
def f_pred(X,w): 
    f =  np.dot(X,w)
    return f

f_pred(np.array([1,-0.15]),np.array([-0.1,0.2]))
```

### Q3: What is the value of the mean square error (MSE) regression loss function $L(w)$ for $w_0=-0.1, w_1=0.2$?

☐ $L$ = 


```python
#Loss MSE
def loss_mse(y_pred,y): 
    n = len(y)
    loss = (1/n)*(np.dot((y_pred-y).T,(y_pred-y)))
    return loss

X = np.ones([n,2])
X[:,1] = data[:,0]
y = data[:,1]
y = y[:,None]
w = np.array([-0.1,0.2])[:,None]

y_pred = f_pred(X,w)
loss = loss_mse(y_pred,y)

print(loss)
```

### Q4: What is the value of the gradient of the MSE regression loss function $\frac{\partial}{\partial w} L(w)$ for $w_0=-0.1, w_1=0.2$?

☐ $\frac{\partial}{\partial w} L$ = 


```python
# gradient function definition
def grad_loss(y_pred,y,X):
    n = len(y)
    grad =  (2/n)*np.dot(X.T,(y_pred-y))
    return grad

gradient_loss = grad_loss(y_pred,y,X)
print(gradient_loss)
```

### Q5: What is the value of the MSE regression loss function $L(w)$ after $200$ gradient descent iterations starting at $w_0=-0.1, w_1=0.2$ and selecting a learning rate $\tau=0.2$?

☐ $L$ = 


```python
#gradient descent function definition
def grad_desc(X, y , w_init=np.array([0,0])[:,None] ,tau=0.01, max_iter=500):

    L_iters = np.zeros([max_iter]) # record the loss values
    w_iters = np.zeros([max_iter,2]) # record the parameter values
    w = w_init # initialization
    for i in range(max_iter): # loop over the iterations
        y_pred = f_pred(X,w) # linear predicition function #YOUR CODE HERE
        grad_f = grad_loss(y_pred,y,X) # gradient of the loss #YOUR CODE HERE
        w = w - tau* grad_f # update rule of gradient descent #YOUR CODE HERE
        L_iters[i] = loss_mse(y_pred,y) # save the current loss value 
        w_iters[i,:] = w[0],w[1] # save the current w value 
        
    return w, L_iters, w_iters


# run gradient descent algorithm 
start = time.time()
w_init = np.array([-0.1,0.2])[:,None]
tau = 0.2
max_iter = 200
w, L_iters, w_iters = grad_desc(X,y,w_init,tau,max_iter)
print('Time=',time.time() - start)
print('Loss=',L_iters[-1])
print(w)


# plot
plt.figure(2)
plt.plot(np.array(range(max_iter)), L_iters)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```

### Q6: What is the value of the MSE regression loss function $L(w)$ computed with scikit-learn?

☐ $L$ = 


```python
# run linear regression with scikit-learn
start = time.time()
lin_reg_sklearn = LinearRegression()
lin_reg_sklearn.fit((X[:,1])[:,None],y) # learn the model parameters #YOUR CODE HERE
print('Time=',time.time() - start)


# compute loss value
w_sklearn = np.zeros([2,1])
w_sklearn[0,0] = lin_reg_sklearn.intercept_
w_sklearn[1,0] = lin_reg_sklearn.coef_
print(w_sklearn)
loss_sklearn = loss_mse(f_pred(X,w_sklearn),y)
print('loss sklearn=',loss_sklearn)
print('loss gradient descent=',L_iters[-1]) 
```

### Q8: What is the value of the new loss function $L_2(w)$ after $200$ gradient descent iterations starting at $w_0=-1.1, w_1=5.2, w_2=4.3$ and selecting a learning rate $\tau=0.2$?

☐ $L_2$ = 

```python
#Loss MSE
def loss_mse(y_pred,y): 
    n = len(y)
    loss = (1/n)*(np.sum((y_pred-y)**4))
    return loss

# gradient function definition
def grad_loss(y_pred,y,X):
    n = len(y)
    grad =  (4/n)*np.dot(X.T,(y_pred-y)**3)
    return grad

#YOUR CODE HERE#YOUR CODE HE# gradient descent function definition
def grad_desc(X, y , w_init=np.array([0,0])[:,None] ,tau=0.01, max_iter=500):

    L_iters = np.zeros([max_iter]) # record the loss values
    w_iters = np.zeros([max_iter,2]) # record the parameter values
    w = w_init # initialization
    for i in range(max_iter): # loop over the iterations
        y_pred = f_pred(X,w) # linear predicition function #YOUR CODE HERE
        grad_f = grad_loss(y_pred,y,X) # gradient of the loss #YOUR CODE HERE
        w = w - tau* grad_f # update rule of gradient descent #YOUR CODE HERE
        L_iters[i] = loss_mse(y_pred,y) # save the current loss value 
        w_iters[i,:] = w[0],w[1] # save the current w value 
        
    return w, L_iters, w_iters

n = data.shape[0]
X = np.ones([n,3])
X[:,1:3] = data[:,0:2]
y = data[:,2]
y = y[:,None]

# run gradient descent algorithm 
start = time.time()
w_init = np.array([-1.1,5.2,4.3])[:,None]
tau = 0.2
max_iter = 200
w, L_iters, w_iters = grad_desc(X,y,w_init,tau,max_iter)
print('Time=',time.time() - start)
print(L_iters[-1])
print(w)


# plot
plt.figure(2)
plt.plot(np.array(range(max_iter)), L_iters)
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.show()
```
